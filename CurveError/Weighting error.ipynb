```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import lmfit as fitter
%matplotlib inline
```

##Fit a sigmoid To a Data set (Unweighted)

```python
mod = fitter.models.ExpressionModel('Xmax * 1 / (1 + exp(-(x - Xo) / b))')  #sigmoidal curve
params =  mod.make_params(Xmax=1, b=0.08, Xo=7)

#Bring in the data
x = pd.read_pickle('Xdata.pkl')
xplot = x.drop_duplicates()
xplot.sort()
y = pd.read_pickle('Ydata.pkl')

#Set initial parameters and fit model
params =  mod.make_params(Xmax=1, Xo = 10.75, b =.9)
yBestFit = mod.fit(y, params, x=x)
yplotBest = mod.eval(yBestFit.params,x=xplot)

plt.plot(x,y, 'o');
plt.plot(xplot, yplotBest, '-' ,linewidth = 4);
```

##Fit sigmoid with weighting toward the upper end of the range (using panda data for weighthing)

```python
mod = fitter.models.ExpressionModel('Xmax * 1 / (1 + exp(-(x - Xo) / b))')  #sigmoidal curve
params =  mod.make_params(Xmax=1, b=0.08, Xo=7)

#Set initial parameters and fit model
params =  mod.make_params(Xmax=1, Xo = 10.75, b =.9)
yBestFit = mod.fit(y, params, x=x)
yplotBest = mod.eval(yBestFit.params,x=xplot)

#calculating weightings
residuals = pd.Series(yBestFit.residual)
Weightings = residuals.where(residuals<0.01,0).copy()

#Recrank fitting with weighting to weighting
yUpperFit = mod.fit(y, params, x=x, weights = Weightings) 
yplotUpper = mod.eval(yUpperFit.params,x=xplot)

plt.plot(x,y, 'o');
plt.plot(xplot, yplotBest, '-' ,linewidth = 4);
plt.plot(xplot, yplotUpper, '--',linewidth = 4)
```
